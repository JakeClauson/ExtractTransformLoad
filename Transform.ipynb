```python
import pandas as pd
import numpy as np
```

```python
accidental_teen_deaths= pd.read_csv("Resources/accidental_teen_deaths.csv")
accidental_child_deaths= pd.read_csv("Resources/accidental_child_deaths.csv")
accidental_deaths= pd.read_csv("Resources/accidental_deaths.csv")
mortality_by_state = pd.read_csv("Resources/Firearm_Mortality_by_State.csv")
registered_firearms = pd.read_csv("Resources/Firearms_2010-2018.csv")
mass_shootings = pd.read_csv("Resources/mass_shooting.csv")
mass_shootings.head()
```

```python
accidental_deaths= accidental_deaths[["State", "Incident Date", "# Killed", "Operations"]]
accidental_child_deaths= accidental_child_deaths[["State", "Incident Date", "# Killed", "Operations"]]
accidental_teen_deaths= accidental_teen_deaths[["State", "Incident Date", "# Killed", "Operations"]]
mass_shootings = mass_shootings[["State", "Incident Date", "# Killed", "# Injured"]]
accidental_deaths.head()
```

```python
accidental_deaths= accidental_deaths.rename(columns={"Operations": "Age"})
accidental_child_deaths= accidental_child_deaths.rename(columns={"Operations": "Age"})
accidental_teen_deaths= accidental_teen_deaths.rename(columns={"Operations": "Age"})

accidental_deaths["Age"]= accidental_deaths["Age"].replace(to_replace = np.nan, value = "Adult")
accidental_child_deaths["Age"]= accidental_child_deaths["Age"].replace(to_replace = np.nan, value = "Child")
accidental_teen_deaths["Age"]= accidental_teen_deaths["Age"].replace(to_replace = np.nan, value = "Teen")
```

```python
joined_table = pd.concat([accidental_deaths, accidental_child_deaths, accidental_teen_deaths])
joined_table
```

```python
joined_table.to_csv(r'Resources\joined_table.csv', index=False)
mass_shootings.to_csv(r'Resources\MASS_SHOOTINGS!.csv', index=False)
```
